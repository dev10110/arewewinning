```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url_confirmed = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
df_confirmed = pd.read_csv(url_confirmed)

url_deaths = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
df_deaths = pd.read_csv(url_deaths)


url_recovered = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
df_recovered = pd.read_csv(url_recovered)
```

```python
df_confirmed
```

```python

df_confirmed2=df_confirmed.drop(columns=['Lat','Long']).groupby('Country/Region').sum().transpose().drop(columns=['Guyana'])
df_deaths2=df_deaths.drop(columns=['Lat','Long']).groupby('Country/Region').sum().transpose().drop(columns=['Guyana'])
df_recovered2=df_recovered.drop(columns=['Lat','Long']).groupby('Country/Region').sum().transpose().drop(columns=['Guyana'])

df_confirmed2.index = pd.to_datetime(df_confirmed2.index)
df_deaths2.index = pd.to_datetime(df_deaths2.index)
df_recovered2.index = pd.to_datetime(df_recovered2.index)
```

```python
df_confirmed2
```

```python

ax = df_confirmed2.plot(logy=True, legend=False)
plt.title('Total Number of Cases')

handles, labels = ax.get_legend_handles_labels()
# sort both labels and handles by total number of cases
labels, handles = zip(*sorted(zip(labels, handles), key=lambda t: df_confirmed2[t[0]][-1], reverse=True))

# only show the top 10 countries in the labels
ax.legend(handles[:10], labels[:10])
```

```python

ax = df_deaths2.plot(logy=True, legend=False)
plt.title('Total Number of Deaths')

handles, labels = ax.get_legend_handles_labels()
# sort both labels and handles by total number of cases
labels, handles = zip(*sorted(zip(labels, handles), key=lambda t: df_deaths2[t[0]][-1], reverse=True))

# only show the top 10 countries in the labels
ax.legend(handles[:10], labels[:10])

```

```python
# plot new reported cases against total number of cases so far
window = 7

df_confirmed_new = df_confirmed2.diff().rolling(f'{window}d').sum()
df_deaths_new = df_deaths2.diff().rolling(f'{window}d').sum()
df_recovered_new = df_recovered2.diff().rolling(f'{window}d').sum()

```

```python
fig = plt.figure()

xaxis = 'confirmed'
yaxis = 'new confirmed'

if xaxis == 'confirmed':
    x = df_confirmed2
if xaxis == 'deaths':
    x = df_deaths2
if xaxis == 'recovered':
    x = df_recovered2

if yaxis == 'confirmed':
    y = df_confirmed2
if yaxis == 'deaths':
    y = df_deaths2
if yaxis == 'recovered':
    y = df_recovered2
    
    
if yaxis == 'new confirmed':
    y = df_confirmed_new
if yaxis == 'new deaths':
    y = df_deaths_new
if yaxis == 'new recovered':
    y = df_recovered_new
  
    

for c in df_confirmed2.columns:
    plt.plot(x[c],y[c], label=c)

    
handles, labels = plt.gca().get_legend_handles_labels()
# sort both labels and handles by total number of cases
labels, handles = zip(*sorted(zip(labels, handles), key=lambda t: y[t[0]][-1], reverse=True))

# only show the top 10 countries in the labels
plt.gca().legend(handles[:10], labels[:10],loc='upper left', bbox_to_anchor= (1.01, 1), ncol=1,
            borderaxespad=0, frameon=True)

plt.yscale('log')
plt.xscale('log')

plt.xlabel(xaxis)
plt.ylabel(f'{yaxis} over last {window} days')

plt.grid(True)
plt.xlim(left=10)
plt.ylim(bottom=1)
```
